## –†–∞—Å—á—ë—Ç—ã –∏ –¥–æ–±–∞–≤–ª–µ–Ω–∏–µ —Ä–µ–∑—É–ª—å—Ç–∞—Ç–æ–≤ –≤ —Ç–∞–±–ª–∏—Ü—É

````python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ —ç—Ç–∞–∂–µ–π –∏ –¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏
def floor_cat (row):
    floor_1 = row['floor']
    floor_last = row['floors_total']
    if  floor_1 == 1:
        return '–ü–µ—Ä–≤—ã–π'
    if floor_1 == floor_last:
        return '–ü–æ—Å–ª–µ–¥–Ω–∏–π'
    return '–î—Ä—É–≥–æ–π'
````

````python
# –°–æ–∑–¥–∞–¥–∏–º —Ñ—É–Ω–∫—Ü–∏—é –∫–∞—Ç–µ–≥–æ—Ä–∏–∑–∞—Ü–∏–∏ —ç—Ç–∞–∂–µ–π –∏ –¥–æ–±–∞–≤–∏–º –Ω–æ–≤—ã–π —Å—Ç–æ–ª–±–µ—Ü —Å –∫–∞—Ç–µ–≥–æ—Ä–∏—è–º–∏
data_house['floor_cat'] = data_house.apply(floor_cat, axis = 1)
data_house.head(2)
````
